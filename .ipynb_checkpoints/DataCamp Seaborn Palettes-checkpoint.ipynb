```python
import seaborn as sns
import matplotlib.pyplot as plt

def create_color_wheel(colors, title, ax):
    ax.pie([1] * len(colors), colors=colors, wedgeprops=dict(width=0.5))
    ax.set_title(title, fontsize=24)
    ax.axis('equal')

# Create a figure with 3 subplots side by side
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Primary Colors
primary_colors = ['red', 'blue', 'yellow']
create_color_wheel(primary_colors, "Primary Colors", ax1)

# Secondary Colors
secondary_colors = ['green', 'orange', 'purple']
create_color_wheel(secondary_colors, "Secondary Colors", ax2)

# Tertiary Colors
tertiary_colors = sns.color_palette("hls", 6)  # 6 evenly spaced colors
create_color_wheel(tertiary_colors, "Tertiary Colors", ax3)

plt.tight_layout()
# plt.savefig('color_wheels.png', dpi=300, bbox_inches='tight')
plt.show()

# print("Color wheels visual has been saved as 'color_wheels.png'.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def create_and_save_visual(data, title, filename):
    plt.figure(figsize=(10, 2))
    plt.imshow([data])
    plt.title(title, fontsize=18)
    plt.axis('off')
    plt.tight_layout()
    # plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()

# Hue
hues = sns.husl_palette(10)
create_and_save_visual(hues, 'Hue', 'hue_spectrum.png')

# Saturation
base_color = sns.color_palette("husl", 1)[0]
saturations = [sns.desaturate(base_color, s) for s in np.linspace(0, 1, 10)]
create_and_save_visual(saturations, 'Saturation', 'saturation_gradient.png')

# Value (Brightness)
values = [tuple(v * c for c in base_color) for v in np.linspace(0, 1, 10)]
create_and_save_visual(values, 'Value (Brightness)', 'value_gradient.png')

# print("Visuals have been saved as separate PNG files.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap

# Set a standard figure size
fig_size = (10, 2)

# Function to create and save a uniform palette plot
def save_palette_plot(colors, title, filename):
    plt.figure(figsize=fig_size)
    data = np.arange(len(colors)).reshape(1, len(colors))
    cmap = ListedColormap(colors)
    plt.imshow(data, cmap=cmap, aspect='auto')
    plt.title(title)
    plt.axis('off')
    # plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()

# Qualitative Palette Example
palette = 'bright'
colors = sns.color_palette(palette)
save_palette_plot(colors, f'Qualitative Palette: {palette.capitalize()}', 'qualitative_palette.png')

# Sequential Palette Example
palette = 'viridis'
cmap = sns.color_palette(palette, as_cmap=True)
data = np.linspace(0, 1, 100).reshape(1, -1)
plt.figure(figsize=fig_size)
plt.imshow(data, cmap=cmap, aspect='auto')
plt.title(f'Sequential Palette: {palette.capitalize()}')
plt.axis('off')
# plt.savefig('sequential_palette.png', dpi=300, bbox_inches='tight')
plt.show()
plt.close()

# Diverging Palette Example
palette = 'coolwarm'
cmap = sns.color_palette(palette, as_cmap=True)
data = np.linspace(-1, 1, 100).reshape(1, -1)
plt.figure(figsize=fig_size)
plt.imshow(data, cmap=cmap, aspect='auto')
plt.title(f'Diverging Palette: {palette.capitalize()}')
plt.axis('off')
plt.colorbar(orientation='horizontal')
# plt.savefig('diverging_palette.png', dpi=300, bbox_inches='tight')
plt.show()
plt.close()

# print("All palette images have been saved.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap

# Set a standard figure size
fig_size = (10, 2)

# Function to create and save a uniform palette plot with white background
def save_palette_plot(colors, title, filename):
    fig, ax = plt.subplots(figsize=fig_size, facecolor='white')
    data = np.arange(len(colors)).reshape(1, len(colors))
    cmap = ListedColormap(colors)
    ax.imshow(data, cmap=cmap, aspect='auto')
    ax.set_title(title, fontsize=18)
    ax.axis('off')
    # plt.savefig(filename, dpi=300, bbox_inches='tight', facecolor='white')
    plt.show()
    plt.close(fig)

# Qualitative Palette Example
palette = 'bright'
colors = sns.color_palette(palette)
save_palette_plot(colors, f'Qualitative Palette: {palette.capitalize()}', 'qualitative_palette.png')

# Sequential Palette Example
palette = 'viridis'
cmap = sns.color_palette(palette, as_cmap=True)
data = np.linspace(0, 1, 100).reshape(1, -1)
fig, ax = plt.subplots(figsize=fig_size, facecolor='white')
ax.imshow(data, cmap=cmap, aspect='auto')
ax.set_title(f'Sequential Palette: {palette.capitalize()}', fontsize=18)
ax.axis('off')
# plt.savefig('sequential_palette.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
plt.close(fig)

# Diverging Palette Example
palette = 'coolwarm'
cmap = sns.color_palette(palette, as_cmap=True)
data = np.linspace(-1, 1, 100).reshape(1, -1)
fig, ax = plt.subplots(figsize=fig_size, facecolor='white')
im = ax.imshow(data, cmap=cmap, aspect='auto')
ax.set_title(f'Diverging Palette: {palette.capitalize()}', fontsize=18)
ax.axis('off')
fig.colorbar(im, orientation='horizontal', ax=ax)
# plt.savefig('diverging_palette.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
plt.close(fig)

# print("All palette images have been saved with white backgrounds.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create and visualize different types of palettes
fig, axs = plt.subplots(4, 1, figsize=(10, 4))

# Qualitative palette
qual_palette = sns.color_palette("pastel", 4)
axs[0].imshow([qual_palette], aspect='auto')
axs[0].set_title("Custom Qualitative Palette", fontsize=20)
axs[0].axis('off')

# Sequential palette
seq_palette = sns.color_palette("Blues", 6)
axs[1].imshow([seq_palette], aspect='auto')
axs[1].set_title("Custom Sequential Palette", fontsize=20)
axs[1].axis('off')

# Diverging palette
div_palette = sns.color_palette("RdBu", 12)
axs[2].imshow([div_palette], aspect='auto')
axs[2].set_title("Custom Diverging Palette", fontsize=20)
axs[2].axis('off')

# HEX codes
hex_palette = sns.color_palette(["#FF6B6B", "#4ECDC4", "#45B7D1", "#FAD02E", "#594F4F"])
axs[3].imshow([hex_palette], aspect='auto')
axs[3].set_title("Custom HEX Palette", fontsize=20)
axs[3].axis('off')

plt.tight_layout()
# plt.savefig('custom1.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create and visualize different types of palettes
fig, axs = plt.subplots(3, 1, figsize=(10, 4))

# Create a default cubehelix palette
default_cube = sns.cubehelix_palette(8)
axs[0].imshow([default_cube], aspect='auto')
axs[0].set_title("Default Cubehelix Palette", fontsize=20)
axs[0].axis('off')

# Create a custom cubehelix palette
custom_cube = sns.cubehelix_palette(8, start=.5, rot=-.75, dark=.3, light=.8)
axs[1].imshow([custom_cube], aspect='auto')
axs[1].set_title("Custom Cubehelix Palette", fontsize=20)
axs[1].axis('off')

# Create a reversed cubehelix palette
reversed_cube = sns.cubehelix_palette(8, start=2, rot=0, dark=0, light=.95, reverse=True)
axs[2].imshow([reversed_cube], aspect='auto')
axs[2].set_title("Reversed Cubehelix Palette", fontsize=20)
axs[2].axis('off')

plt.tight_layout()
# plt.savefig('custom2.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create and visualize different types of diverging palettes
fig, axs = plt.subplots(3, 1, figsize=(10, 4))

# Create a default diverging palette
default_div = sns.diverging_palette(220, 20, n=7)
axs[0].imshow([default_div], aspect='auto')
axs[0].set_title("Default Diverging Palette", fontsize=20)
axs[0].axis('off')

# Create a diverging palette with more saturation
saturated_div = sns.diverging_palette(250, 30, l=65, s=75, n=7)
axs[1].imshow([saturated_div], aspect='auto')
axs[1].set_title("Saturated Diverging Palette", fontsize=20)
axs[1].axis('off')

# Create a diverging palette with a darker center
dark_center_div = sns.diverging_palette(150, 275, s=80, l=55, n=9, center="dark")
axs[2].imshow([dark_center_div], aspect='auto')
axs[2].set_title("Dark Center Diverging Palette", fontsize=20)
axs[2].axis('off')

plt.tight_layout()
# plt.savefig('diverging_palettes.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset('diamonds')

# Using a sequential palette that lightens with increasing quality of diamond color
sns.countplot(data=diamonds, x='color', palette='Blues')  
plt.title('Distribution of Diamond Color Using Sequential Palette', fontsize=15)
plt.xlabel('Color (D is Best, J is Worst)', fontsize=13)
plt.ylabel('Count', fontsize=13)
# plt.savefig('DiamondColorSequential.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Create a discrete cubehelix palette with lighter shades for higher clarity
n_colors = len(diamonds['clarity'].unique())
cube_palette = sns.cubehelix_palette(n_colors, start=.5, rot=-.7, dark=0.3, light=0.8)

# Create the plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=diamonds, x='carat', y='price', hue='clarity', palette=cube_palette, alpha=0.75)
plt.title('Diamond Price vs Carat Weight (Cubehelix Palette for Clarity)', fontsize=15)
plt.xlabel('Carat Weight', fontsize=13)
plt.ylabel('Price (USD)', fontsize=13)
# plt.savefig('DiamondClarityCH.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```
